## Monthly NEX

```python
#### open light curve files & FILTER
## input invalid bins with flux = 0 and flux_error = flux[previous_bin]

## monthly

source_names_monthly = []
UL_bright_ratio_monthly = []

## case zero
nExcess_monthly_zero = []
unc_nExcess_monthly_zero = []
fvar_monthly_zero = []
unc_fvar_monthly_zero = []


for source_name in json_source_names:
    file_name = f'4FGL+{source_name}.json'
    source_data = Read_json_file(file_name, 'monthly', 'fixed')
    source_df = source_data.dataframe
    source_dict = source_data.dictionary
    
    filtered_df, median_df, inputed_df, n_unconstrained = check_and_fill_bins(source_df)
    
    ## UL check
    if len(median_df['flux'].dropna()) != 0:
        bright_ratio = n_unconstrained / len(median_df['flux'].dropna())
    else:
        bright_ratio = 10.0
        
    UL_bright_ratio_monthly.append(bright_ratio)
    
    variability = Estimate_variability(inputed_df)
    sigma_nxs, err_sigma_nxs, fvar, err_fvar = variability.calculate_variability()

    source_names_monthly.append(source_name)

    nExcess_monthly_zero.append(sigma_nxs)
    unc_nExcess_monthly_zero.append(err_sigma_nxs)
    fvar_monthly_zero.append(fvar)
    unc_fvar_monthly_zero.append(err_fvar)
    
df_variability_monthly_zero = pd.DataFrame(list(zip(source_names_monthly, UL_bright_ratio_monthly,
                                                     nExcess_monthly_zero, unc_nExcess_monthly_zero,
                                                     fvar_monthly_zero, unc_fvar_monthly_zero)),
                                            columns =['Source_Name', 'Bright_Ratio', 'NEX_monthly', 'Unc_NEX_monthly',
                                                      'Fvar_monthly', 'Unc_Fvar_monthly'])
```

## Selection of sources 3-sigma monthly variability

```python
table_variability_monthly_zero = Table.from_pandas(df_variability_monthly_zero)
```

```python
## monthly significant 3 sigma

selected_indices = []

for index, row in enumerate(table_variability_monthly_zero):
    
    ## Check if 'Norm_Excess_Var(monthly)' and 'Unc_Norm_Excess_Var(monthly)' are not None and not NaN
    if (row['NEX_monthly'] is not None and not pd.isna(row['NEX_monthly'])) and \
       (row['Unc_NEX_monthly'] is not None and not pd.isna(row['Unc_NEX_monthly'])):
        
        ## 3 sigma criteria
        if row['NEX_monthly'] - 3 * row['Unc_NEX_monthly'] > 0:
            selected_indices.append(index)  # add index to list

if selected_indices:
    table_selected_monthly = table_variability_monthly_zero[selected_indices]
else:
    print("No rows satisfy the condition.")
```

```python
#### open light curve files & FILTER
## input invalid bins with flux = 0 and flux_error = flux[previous_bin]

## 3-days

source_names_3days = []
UL_bright_ratio_3days = []

## case zero
nExcess_3days_zero = []
unc_nExcess_3days_zero = []
fvar_3days_zero = []
unc_fvar_3days_zero = []


for source_name in json_source_names:
    file_name = f'4FGL+{source_name}.json'
    source_data = Read_json_file(file_name, '3-days', 'fixed')
    source_df = source_data.dataframe
    source_dict = source_data.dictionary
    
    filtered_df, median_df, inputed_df, n_unconstrained = check_and_fill_bins(source_df)
    
    ## UL check
    if len(median_df['flux'].dropna()) != 0:
        bright_ratio = n_unconstrained / len(median_df['flux'].dropna())
    else:
        bright_ratio = 10.0
        
    ## 50% cut
    if bright_ratio < 1.0:
        # print(bright_ratio)
        UL_bright_ratio_3days.append(bright_ratio)
        
        variability = Estimate_variability(inputed_df)
        sigma_nxs, err_sigma_nxs, fvar, err_fvar = variability.calculate_variability()
    
        source_names_3days.append(source_name)
        
        nExcess_3days_zero.append(sigma_nxs)
        unc_nExcess_3days_zero.append(err_sigma_nxs)
        fvar_3days_zero.append(fvar)
        unc_fvar_3days_zero.append(err_fvar)
    
df_variability_3days_zero = pd.DataFrame(list(zip(source_names_3days, UL_bright_ratio_3days,
                                                     nExcess_3days_zero, unc_nExcess_3days_zero,
                                                     fvar_3days_zero, unc_fvar_3days_zero)),
                                            columns =['Source_Name', 'Bright_Ratio', 'NEX_3days', 'Unc_NEX_3days',
                                                      'Fvar_3days', 'Unc_Fvar_3days'])
```

## ODR fit

```python
data = selected_sources_df[selected_sources_df['NEX_3days'].notna()]
x_monthly = data['NEX_monthly']
y_3days = data['NEX_3days']
x_err_monthly = data['Unc_NEX_monthly']
y_err_3days = data['Unc_NEX_3days']
```

```python
def power_law_model(B, x):
    return x**(B[0])

def scaled_power_law_model(B, x):
    return B[1] * x**(B[0])
```

```python
from scipy.odr import Model, RealData, ODR

## fit for power law
power_model = Model(power_law_model)
data_power = RealData(x_monthly, y_3days, sx=x_err_monthly, sy=y_err_3days)
odr_power = ODR(data_power, power_model, beta0=[1.0])
output_power = odr_power.run()

## fit for scaled power law
scaled_model = Model(scaled_power_law_model)
data_scaled = RealData(x_monthly, y_3days, sx=x_err_monthly, sy=y_err_3days)
odr_scaled = ODR(data_scaled, scaled_model, beta0=[1.0, 1.0])
output_scaled = odr_scaled.run()
```

```python
## residual variance
residual_variance_power = output_power.res_var
residual_variance_scaled = output_scaled.res_var

print(f"Residual variance Power: {residual_variance_power:.3e}")
print(f"Residual variance Scaled: {residual_variance_scaled:.3e}")

## extract fitted parameters

## power law
a_fit_power = output_power.beta[0]
a_fit_err_power = output_power.sd_beta[0]
print('********* a_fit_err_power = output_power.sd_beta[0] = ', a_fit_err_power)
print(f"Fitted model: y = x^{a_fit_power:.3f} ± {a_fit_err_power:.3f}")

cov_matrix_power = output_power.cov_beta

## scaled power law
a_fit_scaled = output_scaled.beta[0]
a_fit_err_scaled = output_scaled.sd_beta[0]
print('********* a_fit_err_scaled = output_scaled.sd_beta[0] = ', a_fit_err_scaled)
b_fit_scaled = output_scaled.beta[1]
b_fit_err_scaled = output_scaled.sd_beta[1]
print('********* b_fit_err_scaled = output_scaled.sd_beta[1] = ',b_fit_err_scaled)
print(f"Fitted model: y = ({b_fit_scaled:.3f} ± {b_fit_err_scaled:.3f}) * x^({a_fit_scaled:.3f} ± {a_fit_err_scaled:.3f})")

cov_matrix_scaled = output_scaled.cov_beta
```

```python
def fit_free(a, b, x):
    return b * x**a

def fit_fixed(B, x):
    return x**B
```

```python

## create x-range for plotting the fits
x_fit = np.logspace(np.log10(x_monthly.min()), np.log10(x_monthly.max()), len(x_monthly))
y_fit_fixed = fit_fixed(a_fit_power, x_fit)
y_fit_free = fit_free(a_fit_scaled, b_fit_scaled, x_fit)
```

```python
plt.figure(figsize=(8, 6), dpi=300)
plt.errorbar(x_monthly, y_3days, xerr=x_err_monthly, yerr=y_err_3days, fmt='o', markersize=5, alpha=0.5,
             label='Normalized Excess Variance', capsize=2, elinewidth=1, capthick=1.2)

plt.plot(x_fit, y_fit_free, label=f'Fit: $y = ({b_fit_scaled:.3f} ± {b_fit_err_scaled:.3f}) * x^{{{a_fit_scaled:.3f} ± {a_fit_err_scaled:.3f}}}$', color='green', linestyle='--')
plt.plot(x_fit, y_fit_fixed, label=f'Fit (PowerLaw): $y = x^{{{a_fit_power:.3f} ± {a_fit_err_power:.3f}}}$', color='red', linestyle='--')

plt.xlabel('$\sigma_{\mathrm{NXS}}^{2} \pm \mathrm{err}(\sigma_{\mathrm{NXS}}^{2})$ (monthly)', fontsize=14)
plt.ylabel('$\sigma_{\mathrm{NXS}}^{2} \pm \mathrm{err}(\sigma_{\mathrm{NXS}}^{2})$ (3-days)', fontsize=14)
plt.title(f'NXS(3-days) vs. NXS(monthly) - {len(x)} sources', fontsize=16)

plt.xscale('log')
plt.yscale('log')

plt.xlim(1e-2, 1e1)

plt.grid(visible=True, which="both", linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend(fontsize=12)

plt.savefig('plots/selected_fit.png', dpi=300)
plt.show()
```

```python
## we want to keep the previous value of NXS_3days, and then update the NaNs

# Count the number of non-NaN values in NEX_3days before updating
prev_count = selected_sources_df["NEX_3days"].notna().sum()

# Create the updated column using NEX_3days where available, otherwise use the fit
selected_sources_df["NXS_3days_updated"] = np.where(
    selected_sources_df["NEX_3days"].notna(),  
    selected_sources_df["NEX_3days"],  # Keep original values if available
    selected_sources_df["NEX_monthly"] ** a_fit_power  # Apply fit if NaN
)

# Update the flag column: "Y" where we applied the fit, "N" otherwise
selected_sources_df["flag"] = np.where(
    selected_sources_df["NEX_3days"].isna(), "Y", "N"
)

# Count the number of non-NaN values in NXS_3days_updated after updating
updated_count = selected_sources_df["NXS_3days_updated"].notna().sum()

filled_count = updated_count - prev_count

print(f"We previously had {prev_count} estimates for NXS_3days.")
print(f"Now, we have updated it to {updated_count} values, filling in {filled_count} missing estimates using the fit.")
```
